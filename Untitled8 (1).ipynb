```python
Project Title: AI-Based Text-to-Image Generator

Description
Ye project HuggingFace aur Stable Diffusion model ka use karta hai jisme aap text jaise (“A mango on table”) daalte ho aur model us text ke base pe ek image generate karta hai. Ye deep learning ka practical example hai jisme Transformers aur diffusion model ka istemal hota hai. Output image matplotlib se show hoti hai.

Technologies Used:

HuggingFace Diffusers

PyTorch

Transformers

Matplotlib

Google Colab/Jupyter

# STEP 1: Install necessary libraries (if not already installed)
# Run this in a Jupyter cell
!pip install diffusers transformers accelerate scipy safetensors matplotlib

# STEP 2: Import required libraries
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt

# STEP 3: Load the model (with float32 for CPU)
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float32,         # ✅ Required for CPU
    use_safetensors=True
).to("cpu")                             # ✅ Run on CPU

# STEP 4: Generate Image
prompt = "A futuristic city at sunset, cyberpunk style, ultra detailed"
image = pipe(prompt).images[0]

# STEP 5: Show Image
plt.imshow(image)
plt.axis("off")
plt.title("Generated Image")
plt.show()
```
